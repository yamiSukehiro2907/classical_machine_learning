```python
import pandas as pd
df = pd.read_csv('telco_churn.csv')
```

```python
df.head()
```

```python
df.drop(columns=['customerID'] , inplace=True)
```

```python
columns = df.columns
YesNoColumns =[]
for col in columns:
    if(df[col].nunique() > 2):
        continue
    YesNoColumns.append(col)
```

```python
for col in YesNoColumns:
    df[col].replace({"Yes" : 1 , "No": 0 , 1 : 1 , 0 : 0} , inplace=True)
```

```python
df[df.TotalCharges == " "].index
```

```python
df.drop(df[df.TotalCharges == " "].index , inplace=True)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
cat = df.select_dtypes('object')
categorical_columns = cat.columns
categorical_columns
```

```python
dummies = pd.get_dummies(data=cat , drop_first=True , dtype=int)
```

```python
dummies
```

```python
df.drop(columns=categorical_columns , inplace=True)
```

```python
df = pd.concat([df , dummies] , axis=1)
```

```python
df.head()
```

```python
X = df.drop(columns='Churn')
Y = df.Churn
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , train_size=0.7 , random_state=100)
```

```python
numerical_columns = ['MonthlyCharges' , 'TotalCharges' , 'tenure']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.fit_transform(X_test[numerical_columns])
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train , Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test , Y_pred)
```

```python
# Accuracy = no.of correct classification / total classifications
# Accuracy does not properly work with unbiased number of data (unbalanced data)
```

```python
# Sensitivity / Recall = TP / (TP + FN)
# How many actual postivies we got
```

```python
# Specificity = TN / (TN + FP)
# How many actual negatives we got
```

```python
# Precision = TP / (TP + FP)
# How many predicted positive were actually positive
```

```python
# Case-1 
# Cancer
# Actual     Predicted
# Cancer     Not Cancer    FN
# Not Cancer Cancer        FP
# We should not let cancerous people get unonoticed so we should have low FN , so we need high recall
```

```python
# Case-2
# Actual    Predicted
# Spam      Not spam     FN
# Not spam  Spam         FP
# We should not let any important email to go into spam folder so we need low FP , meaning high precision
```

```python
# Case-2
# Actual      Predicted
# Fraud       Valid      FN
# Valid       Fraud      FP
# We should make sure client's experience should not be bad , as well as less number of frauds to be labelled as valid transaction
# We cannot use mean as we want to precision and both recall to be high so we use harmonic mean
```

```python
# High Sensitivity and High Specificity that means model have high classification power
# High Sensitivity and low (1 - specificty) also means model have high classification power
```

```python
# On a ROC-AUC curve if the relation between sensitivity and (1 - specificity) gives us a high area under curve then model is good
```

```python
from sklearn.metrics import roc_auc_score , roc_curve
roc_curve(Y_test , Y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test , Y_pred)
```
