```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score, mean_squared_error
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 100

def analyze_rfe(rfe, X_train, Y_train=None, X_test=None, Y_test=None, feature_names=None, task='auto'):
    """
    Complete RFE analysis - Works for both Classification and Regression
    
    Parameters:
    -----------
    task: 'auto', 'classification', or 'regression'
    """
    
    if feature_names is None:
        feature_names = [f"Feature_{i}" for i in range(X_train.shape[1])]
    
    # Auto-detect task type
    if task == 'auto' and Y_train is not None:
        unique_vals = len(np.unique(Y_train))
        task = 'classification' if unique_vals < 20 else 'regression'
    
    summary_df = pd.DataFrame({
        'Feature': feature_names,
        'Selected': rfe.support_,
        'Ranking': rfe.ranking_
    }).sort_values('Ranking')
    
    # Text Summary
    print("=" * 80)
    print("🔍 RFE ANALYSIS REPORT")
    print("=" * 80)
    print(f"\n📊 OVERVIEW:")
    print(f"   • Total Features: {len(feature_names)}")
    print(f"   • Selected Features: {rfe.n_features_}")
    print(f"   • Eliminated Features: {len(feature_names) - rfe.n_features_}")
    print(f"   • Reduction Rate: {(1 - rfe.n_features_/len(feature_names))*100:.1f}%")
    print(f"   • Task Type: {task.upper()}")
    
    print("\n" + "-" * 80)
    print("✅ SELECTED FEATURES:")
    print("-" * 80)
    selected = summary_df[summary_df['Selected'] == True]
    for i, (idx, row) in enumerate(selected.iterrows(), 1):
        print(f"   {i:2d}. {row['Feature']}")
    
    print("\n" + "-" * 80)
    print("❌ ELIMINATED FEATURES:")
    print("-" * 80)
    eliminated = summary_df[summary_df['Selected'] == False].sort_values('Ranking', ascending=False)
    if len(eliminated) > 0:
        for idx, row in eliminated.iterrows():
            print(f"   Rank {row['Ranking']:2d}: {row['Feature']}")
    
    # Display DataFrame
    print("\n" + "=" * 80)
    print("📋 DETAILED TABLE:")
    print("=" * 80)
    display_df = summary_df.copy()
    display_df['Status'] = display_df['Selected'].map({True: '✅ Selected', False: '❌ Eliminated'})
    display(display_df[['Feature', 'Status', 'Ranking']])
    
    # Visualizations
    fig = plt.figure(figsize=(16, 10))
    
    # Plot 1: Rankings
    ax1 = plt.subplot(2, 2, 1)
    colors = ['#2ecc71' if x else '#e74c3c' for x in summary_df['Selected']]
    y_pos = np.arange(len(summary_df))
    ax1.barh(y_pos, summary_df['Ranking'], color=colors, alpha=0.8, edgecolor='black')
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels(summary_df['Feature'], fontsize=9)
    ax1.set_xlabel('Ranking (1 = Selected)', fontweight='bold')
    ax1.set_title('Feature Ranking', fontweight='bold', fontsize=12)
    ax1.invert_yaxis()
    ax1.grid(axis='x', alpha=0.3)
    
    # Plot 2: Pie chart
    ax2 = plt.subplot(2, 2, 2)
    selected_count = rfe.support_.sum()
    eliminated_count = len(rfe.support_) - selected_count
    ax2.pie([selected_count, eliminated_count],
            labels=[f'Selected\n({selected_count})', f'Eliminated\n({eliminated_count})'],
            autopct='%1.1f%%',
            colors=['#2ecc71', '#e74c3c'],
            explode=(0.05, 0.05))
    ax2.set_title('Selection Distribution', fontweight='bold')
    
    # Plot 3: Histogram
    ax3 = plt.subplot(2, 2, 3)
    ranking_counts = summary_df['Ranking'].value_counts().sort_index()
    bars = ax3.bar(ranking_counts.index, ranking_counts.values, color='steelblue', edgecolor='black')
    ax3.set_xlabel('Ranking', fontweight='bold')
    ax3.set_ylabel('Count', fontweight='bold')
    ax3.set_title('Ranking Distribution', fontweight='bold')
    ax3.grid(axis='y', alpha=0.3)
    
    # Plot 4: Timeline
    ax4 = plt.subplot(2, 2, 4)
    if len(eliminated) > 0:
        elim_order = eliminated.sort_values('Ranking')
        ax4.plot(range(len(elim_order)), elim_order['Ranking'].values,
                marker='o', linewidth=2, markersize=8, color='crimson')
        ax4.fill_between(range(len(elim_order)), elim_order['Ranking'].values, alpha=0.3, color='crimson')
        ax4.set_xlabel('Elimination Order', fontweight='bold')
        ax4.set_ylabel('Ranking', fontweight='bold')
        ax4.set_title('Elimination Timeline', fontweight='bold')
        ax4.grid(True, alpha=0.3)
    
    plt.suptitle('RFE Analysis Dashboard', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Performance comparison if data provided
    if Y_train is not None and X_test is not None and Y_test is not None:
        print("\n" + "=" * 80)
        print("📈 PERFORMANCE COMPARISON")
        print("=" * 80)
        
        X_train_selected = rfe.transform(X_train)
        X_test_selected = rfe.transform(X_test)
        
        if task == 'classification':
            from sklearn.linear_model import LogisticRegression
            model_all = LogisticRegression(max_iter=1000)
            model_all.fit(X_train, Y_train)
            train_acc_all = model_all.score(X_train, Y_train)
            test_acc_all = model_all.score(X_test, Y_test)
            
            model_selected = LogisticRegression(max_iter=1000)
            model_selected.fit(X_train_selected, Y_train)
            train_acc_selected = model_selected.score(X_train_selected, Y_train)
            test_acc_selected = model_selected.score(X_test_selected, Y_test)
            
            print(f"\n🔵 ALL FEATURES ({X_train.shape[1]} features):")
            print(f"   Train Accuracy: {train_acc_all:.4f}")
            print(f"   Test Accuracy:  {test_acc_all:.4f}")
            
            print(f"\n🟢 SELECTED FEATURES ({rfe.n_features_} features):")
            print(f"   Train Accuracy: {train_acc_selected:.4f}")
            print(f"   Test Accuracy:  {test_acc_selected:.4f}")
            
            print(f"\n📊 IMPROVEMENT:")
            print(f"   Accuracy Change: {(test_acc_selected - test_acc_all):+.4f}")
            
        else:  # Regression
            from sklearn.linear_model import LinearRegression
            
            model_all = LinearRegression()
            model_all.fit(X_train, Y_train)
            y_pred_train_all = model_all.predict(X_train)
            y_pred_test_all = model_all.predict(X_test)
            
            train_r2_all = r2_score(Y_train, y_pred_train_all)
            test_r2_all = r2_score(Y_test, y_pred_test_all)
            test_rmse_all = np.sqrt(mean_squared_error(Y_test, y_pred_test_all))
            
            model_selected = LinearRegression()
            model_selected.fit(X_train_selected, Y_train)
            y_pred_train_selected = model_selected.predict(X_train_selected)
            y_pred_test_selected = model_selected.predict(X_test_selected)
            
            train_r2_selected = r2_score(Y_train, y_pred_train_selected)
            test_r2_selected = r2_score(Y_test, y_pred_test_selected)
            test_rmse_selected = np.sqrt(mean_squared_error(Y_test, y_pred_test_selected))
            
            print(f"\n🔵 ALL FEATURES ({X_train.shape[1]} features):")
            print(f"   Train R²: {train_r2_all:.4f}")
            print(f"   Test R²:  {test_r2_all:.4f}")
            print(f"   Test RMSE: {test_rmse_all:.4f}")
            
            print(f"\n🟢 SELECTED FEATURES ({rfe.n_features_} features):")
            print(f"   Train R²: {train_r2_selected:.4f}")
            print(f"   Test R²:  {test_r2_selected:.4f}")
            print(f"   Test RMSE: {test_rmse_selected:.4f}")
            
            print(f"\n📊 IMPROVEMENT:")
            print(f"   R² Change:   {(test_r2_selected - test_r2_all):+.4f}")
            print(f"   RMSE Change: {(test_rmse_selected - test_rmse_all):+.4f}")
        
        print(f"   Feature Reduction: {(1 - rfe.n_features_/X_train.shape[1])*100:.1f}%")
    
    print("\n" + "=" * 80)
    return summary_df
```

```python
# Fit RFE with your regression model
rfe = RFE(lr, n_features_to_select=12)  # lr should be LinearRegression, not LogisticRegression
rfe.fit(X_train, Y_train)

# Run analysis (it will auto-detect regression)
summary = analyze_rfe(rfe, X_train, Y_train, X_test, Y_test, 
                      feature_names=X_train.columns)

# Get selected features
selected_features = X_train.columns[rfe.support_].tolist()
print("\n🎯 Selected Features:", selected_features)
```
