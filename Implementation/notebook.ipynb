```python
categorical_columns = df.select_dtypes(include='object')
categorical_columns
```

```python
numerical_columns = df.select_dtypes(include=['float64' , 'int64']).columns
numerical_columns
```

```python
numerical_columns = numerical_columns[2:]
```

```python
numerical_columns = numerical_columns[:-1]
numerical_columns
```

```python
categorical_dummy =  pd.get_dummies(categorical_columns , drop_first=True , dtype=int)
```

```python
df.drop(columns=categorical_columns.columns , inplace=True)
```

```python
df = pd.concat([df , categorical_dummy] , axis=1)
df.head()
```

```python
X = df.drop(columns='price')
Y = df.price
X
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , train_size=0.7 , random_state=42)
```

```python
X_train[numerical_columns]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.fit_transform(X_test[numerical_columns])
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train , Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test , Y_pred)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(lr , n_features_to_select=12)
rfe.fit(X_train , Y_train)
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[X_train.columns[rfe.support_]]
X_test_rfe = X_test[X_test.columns[rfe.support_]]
```

```python
import statsmodels.api as sm
X_train_sm = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(Y_train, X_train_sm).fit()
print(lr_1.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values , i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'] , 2)

vif = vif.sort_values(by = 'VIF' , ascending=False)
vif
```
